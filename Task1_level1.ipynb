```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# studying data from data file
df = pd.read_csv("data/Dataset .csv")
# View top 5 rows of the dataset
df.head()
```

```python
# Explore the dataset and identify the number of rows and columns.
print("Number of row", df.shape[0])
print("Number of columns", df.shape[1])
```

```python
# Check for missing values in each column and handle them accordingly.
df.isnull().sum()
```

```python

# Checking Null Value by Plotting Heatmap

# Set the plot size
plt.figure(figsize = (10,6))

# Create the figure object
sns.heatmap(df.isnull().corr(), vmin=-1, annot= True)

# Set labels
plt.xlabel('X label', fontsize=11)
plt.ylabel('Y label', fontsize=10)
plt.title('Places of missing values in Dataset', fontsize=12)

plt.show()
```

```python
# Dropping cusines as it contain missing value i.e 9
df = df.dropna(subset=['Cuisines'])
```

```python
# Checking missing values again
df.isna().sum()
```
